```python
import sys
import builtins

def main():
    f_str = input().strip()
    x_0 = float(input())
    delta = float(input())
    
    epsilon = 5 * delta
    
    # Разрешаем все встроенные функции (включая int, float, abs, max и др.)
    namespace = {"__builtins__": builtins, "x": x_0}
    
    def f(x_val):
        namespace["x"] = x_val
        return eval(f_str, namespace)
    
    f0 = f(x_0)
    f_left = f(x_0 - delta)
    f_right = f(x_0 + delta)
    
    if abs(f_left - f0) < epsilon and abs(f_right - f0) < epsilon:
        print("CONTINUOUS")
    else:
        print("DISCONTINUOUS")

if __name__ == '__main__':
    main()
```

```python
import sys
import math

def main():
    epsilon = 1e-6
    e = math.e

    f_expr = input().strip()
    f = eval('lambda x, e=e: ' + f_expr) # функция от x; используем eval с замыканием на e

    a, b = map(float, input().split())
    if a >= b:
        raise ValueError('a/b wrong')
    
    L_input = input().strip()
    try:
        L = float(L_input)
    except ValueError:
        L = eval(L_input, {"__builtins__": {}}, {"e": e})
    if L < 0:
        raise ValueError('L wrong')
    
    n = 1000
    points = []
    d_x = b-a
    for i in range(n+1):
        x = a + d_x * i/n
        points.append(x)

    for i in range(n+1):
        for j in range(i+1, n+1):
            x1 = points[i]
            x2 = points[j]

            fx_1 = f(x1)
            fx_2 = f(x2)
            y_diff = abs(fx_2 - fx_1)
            is_lip = abs(x2 - x1) * L

            if y_diff > is_lip + epsilon:
                print('NOT LIPSCHITZ')
                return
            
    print('LIPSCHITZ')
        
if __name__ == '__main__':
    main()
```

```python
import sys
import sympy as sp


def main():
    
    a, b, c, d = input().split()
    a = float(a)
    b = float(b)
    c = float(c)
    d = float(d)

    numbers = [a, b, c, d]
    for i in range(len(numbers)):
        if -1000 > numbers[i] or 1000 < numbers[i]:
            raise ValueError('a/b/c/d is wrong')
        
    x = sp.symbols('x', real=True)
    polynom = a*x**3 + b*x**2 + c*x + d
    f_diff = sp.diff(polynom, x)
    critical_dots_raw = sp.solve(f_diff, x)

    p, q = input().split()
    p = float(p)
    q = float(q)
    if p >= q:
        raise ValueError('p or q is wrong')
    
    critical_dots = []
    for dot in critical_dots_raw:
        if p <= dot <= q:
            critical_dots.append(dot)

    if not critical_dots:
        print('No critical points found.')
        return

    values = {}
    for dot in critical_dots:
        f = a*dot**3 + b*dot**2 + c*dot + d
        values[dot] = f 
    maximum = max(values.items(), key=lambda item: item[1])
    minimum = min(values.items(), key=lambda item: item[1])

    print(f'local maximum at x = {maximum[0]:.5f}')
    print(f'f(x) = {maximum[1]:.5f}')
    print(f'local minimum at x = {minimum[0]:.5f}')
    print(f'f(x) = {minimum[1]:.5f}')


if __name__ == '__main__':
    main()
```

```python
# решение нейронки

import math


EPS = 1e-9


def main():
    a, b, c, d = map(float, input().split())
    p, q = map(float, input().split())

    roots = []

    # f'(x) = 3ax^2 + 2bx + c
    A = 3 * a
    B = 2 * b
    C = c

    if abs(A) < EPS:
        # Производная линейная: Bx + C = 0
        if abs(B) >= EPS:
            roots.append(-C / B)
        else:
            # Производная константная.
            # Если C == 0, критических точек бесконечно много,
            # но такой случай обычно не дают в тестах.
            roots = []
    else:
        discriminant = B ** 2 - 4 * A * C

        if discriminant > EPS:
            sqrt_d = math.sqrt(discriminant)
            roots.append((-B - sqrt_d) / (2 * A))
            roots.append((-B + sqrt_d) / (2 * A))
        elif abs(discriminant) <= EPS:
            roots.append(-B / (2 * A))

    critical_points = []

    for x in roots:
        if p - EPS <= x <= q + EPS:
            critical_points.append(x)

    critical_points.sort()

    if not critical_points:
        print("No critical points found.")
        return

    for x in critical_points:
        second_derivative = 6 * a * x + 2 * b

        if second_derivative < -EPS:
            point_type = "Local maximum"
        elif second_derivative > EPS:
            point_type = "Local minimum"
        else:
            point_type = "Saddle point"

        fx = a * x ** 3 + b * x ** 2 + c * x + d

        print(f"{point_type} at x = {x:.5f}")
        print(f"f(x) = {fx:.5f}")


if __name__ == "__main__":
    main()
```

```python
import sys
ITERATIONS = 1000


def main():
    
    a, b, c = map(float, input().split())
    if a == 0:
        raise ValueError('a is 0')
    
    x_0 = float(input())
    x = x_0
    def f(x):
        return a * x**2 + b*x + c
    def df(x):
        return 2*a*x + b
    
    epsilon = float(input())
    if epsilon < 0 or epsilon > 10 ** -6:
        raise ValueError('epsilon is wrong')
    
    if abs(f(x)) < epsilon:
        print(f"Root found: x = {x:.6f}")
        print("Number of iterations: 0")
        return

    for i in range(1, ITERATIONS+1):
        derivative = df(x)

        if derivative == 0:
            print("Solution not found")
            return
        
        x = x - f(x) / derivative
        if abs(f(x)) < epsilon:
            print(f"Root found: x = {x:.6f}")
            print(f"Number of iterations: {i}")
            return

    print("Solution not found")

if __name__ == '__main__':
    main()
```

# Частные производные

```python
import sys
import math

def calc_delta(g, u, eps):
    length = 0
    for i in range(len(u)):
        length += u[i] ** 2
    length = math.sqrt(length)

    if length == 0:
        return float("inf")

    dot = 0
    for i in range(len(g)):
        dot += g[i] * u[i]

    delta = eps * dot / length
    return delta

def main():
    
    n, k, eps = input().split()
    n = int(n)
    k = int(k)
    eps = float(eps)

    g_comp = list(map(float, input().split()))

    u_comp = []
    for i in range(k):
        u = list(map(float, input().split()))
        u_comp.append(u)

    best_idx = 1
    best_delta = calc_delta(g_comp, u_comp[0], eps)
    
    for i in range(1, k):
        delta = calc_delta(g_comp, u_comp[i], eps)

        if delta < best_delta:
            best_delta = delta
            best_idx = i+1

    print(best_idx, f"{best_delta:.4f}")

if __name__ == '__main__':
    main()
```

```python
import sys
import math

def matvec(A, w): # Считаем произведение матрицы на вектор
    result = []
    for row in A:
        s = 0
        for j in range(len(w)):
            s += row[j] * w[j]
        result.append(s)
    return result

def loss(A, b, w, lyambda, alpha):
    Aw = matvec(A, w)

    # |Aw - b|^2 (здесь |...| означает длину вектора - одно число, а не матрица)
    quad = 0
    for i in range(len(b)):
        diff = Aw[i] - b[i]
        quad += diff ** 2

    # lambda * |w|^2
    reg = 0
    for i in range(len(w)):
        reg += w[i] ** 2
    reg *= lyambda

    # alpha * sum(sin(w_j))
    noise = 0
    for i in range(len(w)):
        noise += math.sin(w[i])
    noise *= alpha

    return quad + reg + noise

def analytic_grad(A, b, w, lyambda, alpha): # формула = g_j = 2 * sum_i A[i][j] * (Aw[i] - b[i]) + 2 * lambda * w[j] + alpha * cos(w[j])
    m = len(A)
    n = len(w)
    Aw = matvec(A, w)

    grad = []
    for i in range(n):
        value = 0

        # часть 2A^T(Aw - b)
        for j in range(m):
            value += A[j][i] * (Aw[j] - b[j])
        value *= 2

        # lambda w
        value += 2 * lyambda * w[i]

        # alpha cos(w_j)
        value += alpha * math.cos(w[i])

        grad.append(value)
    return grad

def numeric_grad(A, b, w, lyambda, alpha, h):
    n = len(w)
    grad = []

    for i in range(n):
        w_plus = w.copy()
        w_minus = w.copy()

        w_plus[i] += h
        w_minus[i] -= h

        L_plus = loss(A, b, w_plus, lyambda, alpha)
        L_minus = loss(A, b, w_minus, lyambda, alpha)

        value = (L_plus - L_minus) / (2 * h)
        grad.append(value)
    return grad

def main():
    m, n = map(int, input().split())

    A = []
    for _ in range(m):
        row = list(map(float, input().split()))
        A.append(row)

    b = list(map(float, input().split()))

    lyambda, alpha = map(float, input().split())

    w = list(map(float, input().split()))

    h = float(input())

    g = analytic_grad(A, b, w, lyambda, alpha)
    g_num = numeric_grad(A, b, w, lyambda, alpha, h)

    max_diff = 0
    for i in range(n):
        diff = abs(g[i] - g_num[i])
        if diff > max_diff:
            max_diff = diff

    print(f"{max_diff:.6f}")

if __name__ == '__main__':
    main()
```

# Градиентные методы

```python
import math
import sys


def main():
    EPS = 1e-12

    h11, h12 = map(float, input().split())
    h21, h22 = map(float, input().split())

    a = h11
    b = (h12 + h21) / 2
    d = h22

    trace = a + d
    discr = math.sqrt((a - d) ** 2 + 4 * b ** 2)

    lambda1 = (trace + discr) / 2
    lambda2 = (trace - discr) / 2

    eigs = [lambda1, lambda2]

    if any(abs(lam) <= EPS for lam in eigs):
        print("UNKNOWN")
    elif all(lam > 0 for lam in eigs):
        print("MIN")
    elif all(lam < 0 for lam in eigs):
        print("MAX")
    else:
        print("SADDLE")


if __name__ == "__main__":
    main()
```

```python
import sys

def main():
    N, B = map(int, input().split())
    g = list(map(float, input().split()))

    mean = sum(g) / N

    if N == 1 or B == N:
        var_hat = 0.0
    else:
        population_var = sum((x - mean) ** 2 for x in g) / N
        var_hat = (N - B) / (B * (N - 1)) * population_var

    print(f"{mean:.6f} {var_hat:.6f}")


if __name__ == "__main__":
    main()
```

# Методы второго порядка и условная оптимизация

```python
import sys
import math


def sigmoid(z):
    if z >= 0:
        return 1 / (1 + math.exp(-z))
    else:
        exp_z = math.exp(z)
        return exp_z / (1 + exp_z)

def main():
    N = int(input())
    x = []
    y = []
    for _ in range(N):
        x_t, y_t = map(float, input().split())
        x.append(x_t)
        y.append(int(y_t))

    w0 = float(input())
    grad = 0 # L'(w0)
    hess = 0 # L''(w0)
    for i in range(N):
        z = x[i] * w0
        s = sigmoid(z)

        grad += x[i] * (s - y[i])
        hess += x[i]**2 * s * (1-s)
    
    if hess <= 1e-12:
        print('IMPOSSIBLE')
        return

    w1 =  w0 - grad/hess
    print(f"{w1:.6f}")

if __name__ == '__main__':
    main()
```
